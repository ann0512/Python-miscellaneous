```python
#Plot for ùê∏=ùëöùëê2
import numpy as np
m = np.linspace(0,10,11)
print(m)
```

```python
c = 3*10**8
E = m*c**2
print(E)
```

```python
import matplotlib.pyplot as plt
```

```python
fig1 = plt.figure(figsize=(4.6,3))
ax = fig1.add_axes([0,0,1,1])
ax.plot(m,E,color='red',lw=5)
ax.set_xlim(0,10)
ax.set_xlabel('Mass in grams')
ax.set_ylabel('Energy in Joules')
ax.set_title('E=mc^2');
```

Plot this on a logarthimic scale on the y axis. Place a grid along the y axis ticks as well.

```python
fig2 = plt.figure(figsize=(4.6,3))
ax1 = fig2.add_axes([0,0,1,1])
ax1.set_yscale('log')
ax1.plot(m,E,color='red',lw=5)
ax1.set_xlim(0,10)
ax1.set_xlabel('Mass in grams')
ax1.set_ylabel('Energy in Joules')
ax1.grid(which='both',axis='y')
ax1.set_title('E=mc^2');
```


Yield Curve Data

**We've obtained some yeild curve data for you from the [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield). The data shows the interest paid for a US Treasury bond for a certain contract length. The labels list shows the corresponding contract length per index position.


```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

Plot both curves on the same Figure. Add a legend to show which curve corresponds to a certain year.

```python
# Create Figure (empty canvas)
fig=plt.figure()
# Add set of axes to figure
ax1=fig.add_axes([0,0,1,1])
# Plot on that set of axes
ax1.plot(labels,july16_2007,label='july16_2007')
ax1.plot(labels,july16_2020,label='july16_2020')
ax1.legend(loc='center right')
plt.show()
```

**TASK: The legend in the plot above looks a little strange in the middle of the curves. While it is not blocking anything, it would be nicer if it were *outside* the plot.

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(labels,july16_2007,label='july16_2007')
ax1.plot(labels,july16_2020,label='july16_2020')
ax1.legend(loc=(1.05,0.5))
plt.show()
```

**TASK: While the plot above clearly shows how rates fell from 2007 to 2020, putting these on the same plot makes it difficult to discern the rate differences within the same year.

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(12,8))
axes[0].plot(labels,july16_2007)
axes[0].set_title('July 16, 2007')
axes[1].plot(labels,july16_2020)
axes[1].set_title('July 16, 2020');
```

Plot the same using twin axes. While this plot may actually be more confusing than helpful, its a good exercise in Matplotlib control.**

```python
fig,ax1=plt.subplots(figsize=(12,8))
ax1.plot(labels,july16_2007,color='blue',lw=2)
ax1.set_ylabel('2007',fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
plt.yticks(fontsize=15)

ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(4)

ax2=ax1.twinx()
ax2.plot(labels,july16_2020,color='red',lw=2)
ax2.set_ylabel('2020',fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
plt.yticks(fontsize=15)

ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(4)
    
ax1.set_title('July 16th Yield Curves');
```
